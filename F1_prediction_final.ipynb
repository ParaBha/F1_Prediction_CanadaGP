```python
import pandas as pd
```

```python
qualify = pd.read_csv("F1_2025_QualifyingResults.csv")
qualify.head()
```

```python
qualify.columns
```

```python
qualify = qualify[['Track', 'Position','Driver', 'Team', 'Q1', 'Q2', 'Q3']]
qualify.head()
```

```python
raceresult = pd.read_csv("F1_2025_RaceResults.csv")
```

```python
raceresult = raceresult.drop(columns = ['No','Laps'])
raceresult.head()
```

```python

raceresult.head()
```

```python
springresult = pd.read_csv('F1_2025_SprintResults.csv')
springresult.head()
```

```python
df = raceresult.merge(qualify, on = ['Track'], how  = 'left')
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={
    "Position": "Race_Position",
    "Starting Grid": "Grid_Position"
})
```

```python
df.head()
```

```python
def time_to_seconds(t):
    if pd.isna(t) or ":" not in str(t):
        return None
    m, s = t.split(":")
    return int(m)*60 + float(s)
```

```python
for col in ["Q1","Q2","Q3"]:
    df[col] = df[col].apply(time_to_seconds)
```

```python
print(df["Q1"].unique())
```

```python
df['win'] = (df['Position_y'] == 1).astype(int)
```

```python
df['Position_y'] = pd.to_numeric(df['Position_y'], errors='coerce')
df['Podium'] = (df['Position_y'] <= 3).astype(int)
```

```python
df["DNF"] = df["Time/Retired"].str.contains("Retired", na=False).astype(int)
```

```python
df["Fastest_Lap"] = (df["Set Fastest Lap"] == "Yes").astype(int)
```

```python
df
```

```python
driver_stats = df.groupby(["Driver_x","Team_x"]).agg({
    "Points":"sum",
    "win":"sum",
    "Podium":"sum",
    "Position_y":"mean",
    "Grid_Position":"mean",
    "Fastest_Lap":"sum",
    "DNF":"sum"
}).reset_index()
```

```python
driver_stats["Champion"] = (driver_stats["Points"] == driver_stats["Points"].max()).astype(int)
```

```python
driver_stats.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = driver_stats.drop(['Champion',"Driver_x","Team_x"],axis = 1)
Y = driver_stats['Champion']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, random_state = 34, test_size = 0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 34)
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy = 'most_frequent')
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe1 = Pipeline(steps = [
    ('impute',si),
    ('randomForest', rfc)
])
```

```python
pipe1
```

```python
pipe1.fit(x_train,y_train)
```

```python
rfc
```

```python
y_pred = pipe1.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

from sklearn.preprocessing import MinMaxScaler

```python
mms = MinMaxScaler(feature_range = (0,1))
xscale = mms.fit_transform(driver_stats[['Points','Grid_Position','Fastest_Lap']])
xscale
```

```python
# This assigns the scaled values back to the original columns
driver_stats[['Points', 'Grid_Position', 'Fastest_Lap']] = xscale

# View the result to make sure it looks right
print(driver_stats[['Driver_x', 'Points', 'Grid_Position', 'Fastest_Lap']].head())
```

```python
# Check if it worked!
print(driver_stats[['Points', 'Grid_Position', 'Fastest_Lap']].describe().loc[['min', 'max']])
```

```python
# 1. Define your final X and y using the scaled data
X = driver_stats[['Points', 'Grid_Position', 'Fastest_Lap']]
y = driver_stats['Position_y'].fillna(22) # Giving DNFs a last-place rank

# 2. Train the model
# You don't even need the scaler in the pipeline anymore since you did it manually!
from sklearn.ensemble import RandomForestRegressor
model_final = RandomForestRegressor(n_estimators=100, random_state=42)
model_final.fit(X, y)

# 3. View your True Feature Importance
importance = model_final.feature_importances_
feat_imp = pd.Series(importance, index=X.columns)
print("Final Importance Weights:")
print(feat_imp.sort_values(ascending=False))
```

```python
# A hypothetical driver who qualifies P3 (scaled) and has decent points (scaled)
# Format: [[Points, Grid_Position, Fastest_Lap]]
hypothetical_driver = [[0.5, 0.1, 0]] 

prediction = model_final.predict(hypothetical_driver)
print(f"Predicted Finishing Position: {prediction[0]:.2f}")
```

```python
# A driver with Max Points (1.0), Best Grid (0.0), and a Fastest Lap (1.0)
perfect_driver = [[1.0, 0.0, 1.0]] 

win_prediction = model_final.predict(perfect_driver)
print(f"Perfect Driver Predicted Position: {win_prediction[0]:.2f}")
```

```python
# Create a 'Qualifying Efficiency' feature
# This helps the model distinguish between a "fast car/bad driver" vs "slow car/great driver"
driver_stats['Quali_Efficiency'] = driver_stats['Points'] / (driver_stats['Grid_Position'] + 0.1)
driver_stats
```

```python
# Force the model to look deeper into the data
model_final = RandomForestRegressor(n_estimators=100, max_depth=10, min_samples_split=2)
model_final.fit(X, y)
```

```python
# Get predictions for your whole dataset
all_preds = model_final.predict(X)

# Check the standard deviation. If it's near 0, the model is still stuck.
print(f"Prediction Standard Deviation: {all_preds.std():.4f}")
```

```python
# Create a dictionary of actual results from the last race (China 2026)
# If a driver didn't finish, give them 22.
actual_results = {
    'Kimi Antonelli': 1, 
    'George Russell': 2,
    'Lewis Hamilton': 3,
    'Oscar Piastri': 22,
    'Alexander Albon': 22
}

# Apply these unique ranks to your target variable
driver_stats['Target_Pos'] = driver_stats['Driver_x'].map(actual_results).fillna(11)
y = driver_stats['Target_Pos']
```

```python
# Square the Grid Position to penalize starting at the back more heavily
# Since it's scaled 0-1, squaring it makes small differences more 'visible' to the model
driver_stats['Grid_Impact'] = driver_stats['Grid_Position'] ** 2
driver_stats['Grid_Impact']
```

```python
from sklearn.ensemble import RandomForestRegressor

# Increase max_depth and reduce n_estimators to make individual trees more 'opinionated'
model_final = RandomForestRegressor(
    n_estimators=50, 
    max_depth=None, # Allow trees to grow until they find differences
    min_samples_split=2, 
    random_state=42
)

model_final.fit(X, y)
```

```python
# Make sure X_test has all 4 columns used during fit
X_test = driver_stats[['Points', 'Grid_Position', 'Fastest_Lap', 'Grid_Impact']]
driver_stats['Predicted_Rank'] = model_final.predict(X_test)
```

```python
# Format: [[Points, Grid_Position, Fastest_Lap, Grid_Impact]]
# If Grid_Position is 0.0, Grid_Impact (0.0^2) is also 0.0
hypothetical_case = [[1.0, 0.0, 1.0, 0.0]] 

prediction = model_final.predict(hypothetical_case)
print(f"Predicted Position: {prediction[0]:.2f}")
```

```python
def prepare_f1_data(df):
    # Select our Big Three
    X = df[['Points', 'Grid_Position', 'Fastest_Lap']].copy()
    # Create the derived feature automatically
    X['Grid_Impact'] = X['Grid_Position'] ** 2
    return X

# Now use the function for everything
model_final.fit(prepare_f1_data(driver_stats), y)
preds = model_final.predict(prepare_f1_data(driver_stats))
```

```python
# 1. Prepare all 4 features (the "Big Three" + your new Interaction Feature)
X_final = driver_stats[['Points', 'Grid_Position', 'Fastest_Lap']]
X_final['Grid_Impact'] = X_final['Grid_Position'] ** 2

# 2. Add predictions to your dataframe
driver_stats['Model_Predicted_Pos'] = model_final.predict(X_final)

# 3. Display the "Predicted Championship Standings"
leaderboard = driver_stats[['Driver_x', 'Team_x', 'Model_Predicted_Pos']].sort_values(by='Model_Predicted_Pos')
print(leaderboard.head(10))
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
# A scatter plot showing where the model predicted vs where they actually finished
sns.scatterplot(x=y, y=driver_stats['Model_Predicted_Pos'], alpha=0.7)

# The 'Ideal' Line (Where prediction = reality)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')

plt.title("Model Accuracy: Predicted vs. Actual Finishing Position")
plt.xlabel("Actual Position (last race)")
plt.ylabel("Model Predicted Rank")
plt.grid(True)
plt.show()
```

```python
# Save the leaderboard to a CSV file
leaderboard.to_csv('F1_2026_Predicted_Standings.csv', index=False)

print("Leaderboard saved successfully! You can now open it in Excel or Google Sheets.")
```
